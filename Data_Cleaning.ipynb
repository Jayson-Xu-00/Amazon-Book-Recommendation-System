### cleaning the Authors column

```pyspark
from pyspark.sql.functions import col
# Define the regex pattern: strings that start with '[' and end with ']'
pattern = r"^\[.*\]$"
# Filter the DataFrame to find rows where 'authors' does not conform to the pattern
test = df2.filter(~col("authors").rlike(pattern))

# Show the result
test.show()
```

```pyspark
from pyspark.sql.functions import col
# Define the regex pattern: strings that start with '[' and end with ']'
pattern = r"^\[.*\]$"
# Filter the DataFrame to find rows where 'authors' does not conform to the pattern
df2 = df2.filter(col("authors").rlike(pattern))

# Show the result
df2.show()
```

### cleaning the publishedDate column

```pyspark
from pyspark.sql.functions import col

# Define the regex pattern to match YYYY, YYYY-MM, YYYY-MM-DD, and MMM-YY (like Nov-02)
date_pattern = r"^\d{4}(-\d{2})?(-\d{2})?$|^[Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec]-\d{2}$"

# Filter the DataFrame to find rows where 'publishedDate' does not conform to any of the date patterns
df2 = df2.filter(col("publishedDate").rlike(date_pattern))

# Show the result
df2.show()
```

```pyspark
publishedDate_data=df2.select("publishedDate")
publishedDate_data.show()
```

```pyspark
df2 = df2.na.fill({'ratingsCount': 0})
# Show the DataFrame to verify the changes
df2.show()
```

```pyspark
from pyspark.sql.functions import col, count, when
df_missing_values = df2.select([count(when(col(c).isNull(), c)).alias(c) for c in df2.columns])
df_missing_values.show()
```

### cleaning the publisher column

```pyspark
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
import random

# Collect non-missing publishers
non_missing_publishers = df2.filter(col("publisher").isNotNull()).select("publisher").distinct().rdd.flatMap(lambda x: x).collect()

# Define a UDF to replace missing values
@udf(StringType())
def replace_missing(value):
    return value if value is not None else random.choice(non_missing_publishers)

# Apply the UDF to replace nulls in the publisher column
df2 = df2.withColumn("publisher", replace_missing(col("publisher")))

# Show the result
df2.show()
```

### cleaning the categories column

```pyspark
from pyspark.sql.functions import col
# Filter out rows where 'categories' is null
df2 = df2.filter(col("categories").isNotNull())
# Show the result
df2.show()
```

```pyspark
from pyspark.sql.functions import col, count, when
df_missing_values = df2.select([count(when(col(c).isNull(), c)).alias(c) for c in df2.columns])
df_missing_values.show()
```

```pyspark
row_count = df2.count()
print("Number of rows:", row_count)
```

```pyspark
df2.coalesce(1).write.csv('s3://finalprojectbig/output/cleaned_books_data.csv', header=True, mode='overwrite')
```
